```python
import pandas as pd
import json
file_path = 'C:/Users/boyko/Downloads/dataset_crawler-google-places_2026-03-26_17-18-47-182 (1).json'

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)
df = pd.DataFrame(data)

columns_to_keep = ['title', 'stars', 'text', 'textTranslated', 'publishedAtDate']

# Filtrujemy tabelę
df_clean = df[columns_to_keep].copy()

# usuwamy te recenzje, w których użytkownik po prostu przyznał gwiazdki, ale nic nie napisał
df_clean = df_clean.dropna(subset=['text'])

print(f"Opinie zawierające rzeczywisty tekst do analizy: {len(df_clean)}\n")
display(df_clean.head())
```

```python
#  Tworzymy kolumnę z tonacją opartą na gwiazdkach
def get_sentiment(stars):
    if stars <= 2:
        return 'Negative'
    elif stars == 3:
        return 'Neutral'
    else:
        return 'Positive'

df_clean['sentiment'] = df_clean['stars'].apply(get_sentiment)


df_clean['text_lower'] = df_clean['text'].str.lower()

def categorize_complaint(text):

    categories = {
        'Service & Queues (Obsługa i kolejki)': ['obsług', 'pan', 'pani', 'kasjer', 'kolejk', 'czeka', 'niemil', 'chamsk', 'woln'],
        'Food & Coffee (Jedzenie i kawa)': ['hot', 'dog', 'kaw', 'zimn', 'stare', 'nieswiez', 'jedzen', 'parówk'],
        'Prices & Scam (Ceny i oszustwa)': ['drog', 'cen', 'paragon', 'oszust', 'grosz', 'promocj', 'etykiet'],
        'Cleanliness (Czystość)': ['brud', 'śmierdz', 'zapach', 'bałagan'],
        'Working Hours/Closed (Godziny otwarcia/Dni zamknięcia)': ['zamknięt', 'otwart', 'godzin']
    }
    
    found_categories = []
    for category, keywords in categories.items():
        # Jeśli w recenzji znajduje się choćby jedno słowo z tej kategorii
        if any(keyword in text for keyword in keywords):
            found_categories.append(category)
            
    # Jeśli nic nie znaleźli
    if not found_categories:
        return 'Other (Інше)'
        
    # Jeśli opinia mieści się w kilku kategoriach, łączymy je
    return ' + '.join(found_categories)

#  Analizujemy WYŁĄCZNIE negatywne opinie
negative_reviews = df_clean[df_clean['sentiment'] == 'Negative'].copy()
negative_reviews['complaint_category'] = negative_reviews['text_lower'].apply(categorize_complaint)

print(" Number of negative reviews:", len(negative_reviews))
print("\n Top Reasons for Complaints in Żabki, Lublin:")
display(negative_reviews['complaint_category'].value_counts().head(10).to_frame(name='Liczba skarg'))
```

```python
import plotly.express as px

# (przedstawiamy 10 najczęstszych skarg)
top_complaints = negative_reviews['complaint_category'].value_counts().reset_index()
top_complaints.columns = ['Complaint category', 'Quantity']
top_complaints = top_complaints.head(10)


fig = px.bar(top_complaints, 
             x='Quantity', 
             y='Complaint category', 
             orientation='h',
             title=' Top 10 Reasons for Negative Reviews in Żabki (Lublin)',
             color='Quantity',
             color_continuous_scale='Reds', 
             text_auto=True)


fig.update_layout(yaxis={'categoryorder':'total ascending'}) 

fig.show()
```
